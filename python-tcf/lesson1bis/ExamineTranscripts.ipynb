```python
# initialize counters
my_ects = { 
    'MAT' : 0,
    'FIS' : 0,
    'CHIM': 0,
    'LIN' : 0,
    'INF' : 0,
    'TOT' : 0
}

avg_grades = 0.
i_line, number_cum_laude, tot_ects_with_grades = 0, 0, 0
```

Ripassiamo qui l'istruzione `if`, il ciclo `for` e vediamo esempi di **operazioni su stringhe** molto usate in Python:
* [Il ciclo for](03_08.ipynb) 
* [Il blocco if](03_05.ipynb)
* [Troncamento di liste e stringhe](05_05.ipynb)

```python
with open('./transcripts2.txt') as transc :
    lines = transc.readlines()
    for line in lines :
        
        # print (line.strip())
        if (line.find('SSD') == 0) :  # line should begin with 'SSD'

            ssd = line[line.find(' ')+1:line.find('/')]    # line[a:b] -> da a - b
            
            nline = lines[i_line+1]    # next line
            cfu = int(nline[nline.find(' ')+1:])

            my_ects['TOT'] += cfu
            if (ssd in my_ects.keys()) : my_ects[ssd] += cfu
            
            nnline = lines[i_line+2]   # next-to-next line
            if (nnline.find('/30') > 0) :   # check whether this exam has grades
                
                grades = nnline[nnline.find(' ')+1:nnline.find('/')]    
                if (grades.find('L') > 0) :   # check "cum laude"
                    grades = 30;    number_cum_laude += 1
                else: 
                    grades = int(grades)    
                avg_grades += cfu*grades
                tot_ects_with_grades += cfu
                
        i_line += 1        
```

```python
ok_word = ' '  
ok_to_graduate = True
for exam_type, ects in (my_ects.items()) : 
  ok_this_exam_type = (ects >= minimum_ects[exam_type])
  ok_word = ' ' 
  if (not ok_this_exam_type) : ok_word = '   <--- NON SUPERATO'
  print ('Crediti di tipo ',exam_type,' : ',ects,'/',minimum_ects[exam_type],' necessari',ok_word, sep='')  
  ok_to_graduate = (ok_to_graduate and ok_this_exam_type)

if (not ok_to_graduate) : ok_word = ' non '
print ('\nLo studente',ok_word,'si può laureare',sep='')
print (f'Media voti: {avg_grades/tot_ects_with_grades:.2f}/30')
print ('Numero di lodi:', number_cum_laude)
```
